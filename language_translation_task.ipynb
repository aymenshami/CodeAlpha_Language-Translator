```python
translator = Translator()

def translate_text(text, src_lang, tgt_lang):
    try:

        src_code = list(LANGUAGES.keys())[list(LANGUAGES.values()).index(src_lang)]
        tgt_code = list(LANGUAGES.keys())[list(LANGUAGES.values()).index(tgt_lang)]

        result = translator.translate(text, src=src_code, dest=tgt_code)
        return result.text
    except Exception as e:
        return f"Error: {str(e)}"

langs = list(LANGUAGES.values())

iface = gr.Interface(
    fn=translate_text,
    inputs=[
        gr.Textbox(lines=5, label="Source Text"),
        gr.Dropdown(langs, value="english", label="Source Language"),
        gr.Dropdown(langs, value="arabic", label="Target Language")
    ],
    outputs=gr.Textbox(lines=5, label="Translated Text"),
    title="Language Translator",
    description="A simple language translator using Googletrans + Gradio"
)

iface.launch()
```
